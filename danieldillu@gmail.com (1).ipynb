# **2. Problem Statement**
---
The high cost of new automobiles has made them unaffordable for many people, resulting in an increase in the sales of used cars globally. However, accurately determining the value of a used car can be challenging, requiring a reliable prediction system that considers various features. While some websites claim to offer this service, their prediction methods may not be entirely reliable due to differences in models and systems. Therefore, determining the actual market value of a used car before buying or selling one is crucial.

To address this issue, a Data Science Project has two main objectives. </br>The first objective is to determine the value of used cars by analyzing their characteristics and past sales prices. The second objective is to gain a deeper understanding of the critical factors that influence used vehicle pricing. This will involve identifying trends in car sales and the broader automotive industry while examining the correlation between the price of a car and various factors, such as its year of manufacture and mileage.

With these objectives in mind, a thorough and insightful analysis of the used car market can be conducted.

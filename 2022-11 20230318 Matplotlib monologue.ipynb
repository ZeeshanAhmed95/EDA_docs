```python
import numpy as np
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
# actual plotting code
plt.plot(x,y, color = 'red')

# aux
plt.xlabel('my x values')
plt.ylabel('my y values')
plt.title('MY first Chart')

plt.show()
```

```python
print(x)
```

```python
y1 = x**2
y2 = x**3
```

```python
# actual plotting code
plt.plot(x,y1, color = 'red', label = 'sqr')
plt.plot(x,y2, color = 'blue', label = 'cube')

# aux
plt.xlabel('my x values')
plt.ylabel('my y values')
plt.title('MY second Chart')
plt.legend()

plt.show()
```

```python
plt.subplot(1, 2, 1) # rows and cols and which plot
plt.plot(x,y1, color = 'red', label = 'sqr')
plt.xlabel('my x values')
plt.ylabel('my y values')
plt.title('MY second Chart')
plt.legend()


plt.subplot(1,2,2) 
plt.plot(x,y2, color = 'blue', label = 'cube')
plt.xlabel('my x values')
plt.ylabel('my y values')
plt.title('MY second Chart')
plt.legend()

plt.show()
```

```python
paper = plt.figure(figsize = [15,7]) # canvas

ax1 = paper.add_axes([0.1,0.1,0.8,0.8]) #left,#bot,#legth,#height
ax1.plot(x,y)
ax1.set_xlabel('my x variable')
ax1.set_ylabel('my y variable')
ax1.set_title('my chart')


ax2 = paper.add_axes([0.2,0.5,0.2,0.2])
ax2.plot(x,np.sqrt(x))
ax2.set_xlabel('x value')
ax2.set_ylabel('sqrt')
ax2.set_title('small chart')

plt.show()
```
